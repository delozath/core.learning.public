# Distancia Minkowski y generalización de métricas de agrupamiento

```python
# Carga bibliotecas
import numpy             as np
import matplotlib.pyplot as plt

from sklearn import datasets
%matplotlib inline
```

```python

def dist_eucl_pw(X,C):
    """
     @func: Mide la distancia eclidiana vector a vector entre dos conjuntos de datos
     @param: X (muestras,rasgos)
     @param: C (muestras,rasgos)
    """
    D  = X[:,:,np.newaxis] - np.transpose( C[np.newaxis],(0,2,1) )
    D  = np.transpose(D,(0,2,1))
    D  = np.power(D,2).sum(axis=-1)
    return D.copy()


def dist_minkowski(X,C,p):
    """
      @func: Mide la distancia Minkowski vector a vector entre dos conjuntos
      @param: p numero real
      @param: X (muestras,rasgos)
      @param: C (muestras,rasgos)
    """
    D  = X[:,:,np.newaxis] - np.transpose( C[np.newaxis],(0,2,1) )
    D  = np.abs(  np.transpose(D,(0,2,1))  )
    D  = np.power(D,p).sum(axis=-1)
    return D.copy()

def dist_mahalanobis(X,C):
    """
      @func: Mide la distancia Mahalanobis vector a vector entre dos conjuntos
      @param: p numero real
      @param: X (muestras,rasgos)
      @param: C (muestras,rasgos)
    """
    D  = X[:,:,np.newaxis] - np.transpose( C[np.newaxis],(0,2,1) )
    D  = np.transpose(D,(0,2,1))
    
    sort = []
    for i,d in enumerate(D):
        C  = np.linalg.inv( np.cov(d.T) )
        dp = np.dot(d ,C.T)
        dp = np.dot(d ,dp.T )
        dp = np.diag(dp)
        
        sort.append( dp )
    
    return np.array(sort).copy()
        
```

## Generar vectores de datos para ejemplos

```python
# Blobs de distinta varianza
# Se generan vectore X0 y X1
N = 400
Xblobs, Lblobs   = datasets.make_blobs(n_samples=N, cluster_std=[1.0, 2.5, 0.5],
                                       random_state=170)

Xblobsv, Lblobsv = datasets.make_blobs(n_samples=N, random_state=170)
Xblobsv          = np.dot(Xblobsv, [[1.6, -2.8], [-2.2, 1.8]])


X0   = Xblobs [Lblobs ==1]
X1   = Xblobsv[Lblobsv==2]

plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal', 'datalim')
plt.plot(X0[:,0],X0[:,1],'ok',markersize=10,alpha=.8)
plt.plot(X1[:,0],X1[:,1],'or',markersize=10,alpha=.8)
plt.grid(True)
```

### Agrupamiento aglutinante utilizando distancia euclidiana

```python
from sklearn.cluster import AgglomerativeClustering as agg
```

```python
# Crear un grupo unico de datos fucionando X0 y X1
X = np.concatenate( (X0,X1),axis=0 )
```

```python
D = dist_eucl_pw(X,X)
```

```python
model = agg( affinity='precomputed', compute_full_tree='auto',
             linkage='average', n_clusters=2 )
model.fit(D)
```

```python
mask = model.labels_

plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal', 'datalim')
plt.plot(X0[:,0],X0[:,1],'o',color='orange',markersize=12)
plt.plot(X1[:,0],X1[:,1],'o',color='lime'  ,markersize=12)

plt.plot(X[mask==0,0],X[mask==0,1],'ok',markersize=7)
plt.plot(X[mask==1,0],X[mask==1,1],'or',markersize=7)
plt.plot(X[mask==2,0],X[mask==2,1],'om',markersize=7)
plt.plot(X[mask==3,0],X[mask==3,1],'oc',markersize=7)

plt.grid(True)
```

### Agrupamiento aglutinante utilizando distancia Mahalanobis

```python
D = dist_mahalanobis(X,X)
```

```python
model = agg( affinity='precomputed', compute_full_tree='auto',
             linkage='average', n_clusters=3 )
model.fit(D)
```

```python
mask = model.labels_

plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal', 'datalim')
plt.plot(X0[:,0],X0[:,1],'o',color='orange',markersize=12)
plt.plot(X1[:,0],X1[:,1],'o',color='lime'  ,markersize=12)

plt.plot(X[mask==0,0],X[mask==0,1],'ok',markersize=7)
plt.plot(X[mask==1,0],X[mask==1,1],'or',markersize=7)
plt.plot(X[mask==2,0],X[mask==2,1],'om',markersize=7)
plt.plot(X[mask==3,0],X[mask==3,1],'oc',markersize=7)

plt.grid(True)
```

### Agrupamiento aglutinante utilizando distancia Minkowski

```python
D = dist_minkowski(X,X,np.pi/4)
```

```python
model = agg( affinity='precomputed', compute_full_tree='auto',
             linkage='average', n_clusters=2 )
model.fit(D)
```

```python
mask = model.labels_

plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal', 'datalim')
plt.plot(X0[:,0],X0[:,1],'o',color='orange',markersize=12)
plt.plot(X1[:,0],X1[:,1],'o',color='lime'  ,markersize=12)

plt.plot(X[mask==0,0],X[mask==0,1],'ok',markersize=7)
plt.plot(X[mask==1,0],X[mask==1,1],'or',markersize=7)
plt.plot(X[mask==2,0],X[mask==2,1],'om',markersize=7)
plt.plot(X[mask==3,0],X[mask==3,1],'oc',markersize=7)

plt.grid(True)
```
