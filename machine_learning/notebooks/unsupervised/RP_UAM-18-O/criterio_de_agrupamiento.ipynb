## **Idea general del agrupamiento **

```
# Animales y rasgos [grupo,progenie,pulmones,ecosistema]
#   grupo:      0 mamifero, 1 ave, 2 reptil, 3 pez
#   progenie:   0 vivipara, 1 ovipara
#   pulmones:   0 si, 1 no
#   ecosistema: 0 aerobio, 1 anaerobio, 2 mixto
animales = {'borrego':[0,0,0,0], 'perro':[0,0,0,0], 'gato':[0,0,0,0],
            'cuervo':[1,1,0,0], 'garza':[1,1,0,0], 'cobra':[2,1,0,0],
            'lagartija':[2,1,0,0], 'dorado':[3,1,1,1], 'tilapia':[3,1,1,1],
            'tiburon':[3,0,1,1], 'rana':[2,1,0,2]}
```

```
# Distintos agrupamientos
criterio = ('progenie','pulmones','ecosistema')
for crit, opciones in zip((0,1,2),(2,2,3)):
  print(criterio[crit] + ':')
  for opcion in range(0,opciones):
    print('(' + str(opcion) + ') ')
    for animal in animales:
      if animales[animal][crit+1] == opcion:
        print('  ' + animal)    
  print(' ')      
```

```
# Agrupamiento mixto
for progenie in (0,1):
  for pulmones in (0,1):
    print('({0:d},{1:d})'.format(progenie,pulmones))
    for animal in animales:
      if animales[animal][1] == progenie and animales[animal][2] == pulmones:
        print('  ' + animal)   
```
