# Plano de Laplace 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def hs_1():
    a = np.linspace(-1, 1, 100)
    w = np.linspace(-1, 1, 100)
    a, w = np.meshgrid(a, w)
    #
    Hs = 1 / (a + 1j*w + 1)
    return Hs
```

```python
def plot3D(Hs):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = Hs.real
    y =Hs.imag
    z = -20*np.log10(np.absolute(Hs))
    surface = ax.plot_surface(x, y, z, cmap='viridis')
    #
    plt.show()
```

```python
plot3D(hs_1())
```

```python
def func_hs(func):
    x = np.linspace(-50, 50, 400)
    a, w = np.meshgrid(x, x)
    #
    Hs = func(a, w)
    return Hs

def plot3D(Hs):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = Hs.real
    y = Hs.imag
    z = 20*np.log10(np.absolute(Hs))
    surface = ax.plot_surface(x, y, z, cmap='viridis')
    #
    plt.show()
#
hs_pb_o1 = lambda a, w: 1 / (a + 1j*w + 1)
#
hs = func_hs(hs_pb_o1)
plot3D(hs)
```

```python
def hs_estable(Hs):
    x = Hs.real
    y = Hs.imag
    z = 20*np.log10(np.absolute(Hs))
    #
    fig, ax = plt.subplots()
    #
    c = ax.pcolormesh(x, y, z, cmap='RdBu')
    ax.set_title('pcolormesh')
    # set the limits of the plot to the limits of the data
    #ax.axis([x.min(), x.max(), y.min(), y.max()])
    fig.colorbar(c, ax=ax)
    #
    plt.show()
```

```python
hs_pb_o1 = lambda a, w: 5 / (a + 1j*w + 1)
hs_pa_o1 = lambda a, w: (a + 1j*w) / (a + 1j*w + 35)
#
hs = func_hs(hs_pa_o1)
hs_estable(hs)
```
