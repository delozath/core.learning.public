# Estructuras de datos

## Tuplas

Características:
 - Colección de datos
 - Datos multivariados
 - Inmutable -> Una vez que se crea no se puede modificar

```python
# definición explicita
x = tuple([1, 2, 'd'])
x
```

```python
# por valores
x = (1, 3, '4', 0.1, 3, -4, 1)
x
```

```python
# selección elemento único
x[3]
```

```python
# slice
x[1:3]
```

```python
# Último elemento
x[-1]
```

```python
# Slice sin el último elemento
x[0:-1]
```

```python
# Slice sin el último notación compacta
x[:-1]
```

```python
x.index(1)
```

```python
x.count(1)
```

```python
del(x[0])
```
